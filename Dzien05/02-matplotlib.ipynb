```python
import numpy as np
import matplotlib.pyplot as plt
```

### Wykres punktowy

```python
# 1,1  2,5  3,8   4,10
#plt.plot([1,2,3,4], [1,5,8,10], "s", c='purple')
plt.plot([1,2,3,4], [1,5,8,10], "r^")
plt.grid()
plt.title("Tytu≈Ç wykresu")
plt.axis([0,5,0,15]) # x1, x2, y1, y2
# kolor czerwony - FF0000
plt.show()
```

```python
x = np.arange(0, 5.2, 0.2)
# zmiana wielkosci wykresu
plt.figure(figsize=(10,6))
plt.plot(x, x, "g--")
plt.plot(x, 1/2*x**2, "r")
plt.plot(x, 1/8*x**3, "k-.")
plt.text(0, 12, "Tekst na wykresie", fontsize=30, color="orange")
plt.annotate("Adnotacja na wykresie", [2,2], [3,10], arrowprops={
    "linewidth" : 1, "arrowstyle" : "<->"
})
plt.show()
```

```python
#plt.plot(x, x, "g--")
#plt.plot(x, 1/2*x**2, "r")
#plt.plot(x, 1/8*x**3, "k-.")
plt.plot(x,x,"g--",   x,1/2*x**2,"r",   x,1/8*x**3,"k-.")
plt.show()
```

### Wykres rozrzutu

```python
np.random.seed(0)
N = 100
x = 6*np.random.randn(N) - 3
y = 2*x + np.random.randn(N)*5
```

```python
plt.figure(figsize=(8,8))
plt.grid(linestyle='--', color='blue')
plt.scatter(x, y, color='red', marker="s")
#plt.plot(x, y)

plt.show()
```

### Wykres s≈Çupkowy

```python
countries = ["USA", "Rosja", "Kanada"]
m_bronze = np.array([12, 10, 8])
m_silver = np.array([4,2,8])
m_gold = np.array([2, 2, 1])

plt.bar(countries, m_bronze, color='brown', width=.5, label="Bronze")
plt.bar(countries, m_silver, color='silver', width=.5, bottom=m_bronze, label="Silver")
plt.bar(countries, m_gold, color='gold', width=.5, bottom=m_bronze+m_silver, label="Gold")
plt.legend(loc="upper right", ncol=2)

# zwiekszanie rozmiaru wykresu po u≈ºyciu metod rysujƒÖcych
plt.gcf().set_size_inches(10,7)

# zapis wykresu do pliku graficznego
plt.savefig("wykres.png", dpi=150)
plt.savefig("wykres.jpg", dpi=150, quality=30)
plt.savefig("wykres.tif")

plt.show()
```

```python
x = np.array([1,5,9])
countries = ["USA", "Rosja", "Kanada"]
m_bronze = np.array([12, 10, 8])
m_silver = np.array([4,2,8])
m_gold = np.array([2, 2, 1])

plt.bar(x, m_bronze, color='brown', width=.5, label="Bronze")
plt.bar(x+0.5, m_silver, color='silver', width=.5, label="Silver")
plt.bar(x+1, m_gold, color='gold', width=.5, label="Gold")
plt.legend(loc="upper right", ncol=2)

# zwiekszanie rozmiaru wykresu po u≈ºyciu metod rysujƒÖcych
plt.gcf().set_size_inches(10,7)

# modyfikacja osi X
plt.xticks(x+0.5, countries)
# modyfikacja osi Y
plt.yticks(range(0,16,5))

plt.grid(axis="y")

plt.show()
```

### Wykres ko≈Çowy (pie chart)

```python
names = ["Ja≈õ", "Gosia", "Zuzia", "Marek"]
toys = [10, 7, 17,6]
colors = ["#FF0033", "blue", "yellowgreen", "lightskyblue"]

plt.figure(figsize=(10,7))

plt.pie(toys, labels=names, colors=colors, autopct="%.1f %%", shadow=True, explode=[0.2, 0, 0.1, 0])
plt.show()
```
