```python
import numpy as np
import matplotlib.pyplot as plt
```

### Wykres punktowy

```python
# 1,1  2,5  3,8   4,10
#plt.plot([1,2,3,4], [1,5,8,10], "s", c='purple')
plt.plot([1,2,3,4], [1,5,8,10], "r^")
plt.grid()
plt.title("Tytu≈Ç wykresu")
plt.axis([0,5,0,15]) # x1, x2, y1, y2
# kolor czerwony - FF0000
plt.show()
```

```python
x = np.arange(0, 5.2, 0.2)
# zmiana wielkosci wykresu
plt.figure(figsize=(10,6))
plt.plot(x, x, "g--")
plt.plot(x, 1/2*x**2, "r")
plt.plot(x, 1/8*x**3, "k-.")
plt.text(0, 12, "Tekst na wykresie", fontsize=30, color="orange")
plt.annotate("Adnotacja na wykresie", [2,2], [3,10], arrowprops={
    "linewidth" : 1, "arrowstyle" : "<->"
})
plt.show()
```

```python
#plt.plot(x, x, "g--")
#plt.plot(x, 1/2*x**2, "r")
#plt.plot(x, 1/8*x**3, "k-.")
plt.plot(x,x,"g--",   x,1/2*x**2,"r",   x,1/8*x**3,"k-.")
plt.show()
```

### Wykres rozrzutu

```python
np.random.seed(0)
N = 100
x = 6*np.random.randn(N) - 3
y = 2*x + np.random.randn(N)*5
```

```python
plt.figure(figsize=(8,8))
plt.grid(linestyle='--', color='blue')
plt.scatter(x, y, color='red', marker="s")
#plt.plot(x, y)

plt.show()
```
