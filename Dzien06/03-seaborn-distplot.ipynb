```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("http://51.91.120.89/extras/gapminderDataFiveYear.csv")
df
```

```python
sns.distplot(df.lifeExp, kde=False, bins=30)
```

```python
sns.distplot(df.query(" continent=='Africa' ").lifeExp, kde=True, bins=30, label="Africa")
sns.distplot(df.query(" continent=='Europe' ").lifeExp, kde=True, bins=30, label="Europe")
plt.legend()
plt.show()
```

```python
sns.displot(df.query(" continent=='Africa' ").lifeExp, kde=True, bins=30, label="Africa")
sns.displot(df.query(" continent=='Europe' ").lifeExp, kde=True, bins=30, label="Europe")
plt.legend()
plt.show()
```

```python
sns.histplot(df.query(" continent=='Africa' ").lifeExp, kde=False, bins=10, label="Africa", color="green")
sns.histplot(df.query(" continent=='Europe' ").lifeExp, kde=True, bins=30, label="Europe", color="blue")
plt.legend()
plt.show()
```
