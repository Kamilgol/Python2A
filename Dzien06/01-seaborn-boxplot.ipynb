```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("http://51.91.120.89/extras/pokemon.csv")
df
```

```python
plt.figure(figsize=(12,14))
sns.boxplot(data=df, x="Attack", y="Type 1", orient="h", color="r", whis=1)
```

```python
plt.figure(figsize=(21,11))
sns.boxplot(data=df, y="Attack", x="Type 1", orient="v", color="r", whis=1, hue="Legendary")
```
