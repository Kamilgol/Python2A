```python
import requests
import bs4
```

```python
# pobranie HTMLa z wskazanego URLa
response = requests.get("https://www.imdb.com/chart/top")
```

```python
# przekształcenie HTMLa do DOM (Document Object Model)
dom = bs4.BeautifulSoup(response.text)
```

```python
tbody = dom.find("tbody", class_="lister-list") # pobranie zawartości tabeli wg znacznika TBODY i nazwy klasy
```

```python
rows = tbody.find_all("tr") # ekstrakcja wierszy z tabeli
```

```python
#rows[0].find("td", class_="titleColumn").find("a").get("href")
rows[0].find("td", class_="posterColumn").find("img").get("src")
```

```python
result = []
for row in rows:
    rank = row.find("span").get("data-value")
    link = row.find("td", class_="titleColumn").find("a").get("href")
    title = row.find("td", class_="titleColumn").find("a").text
    year = row.find("td", class_="titleColumn").find("span").text[1:-1]
    rating = row.find("td", class_="ratingColumn").text.strip()
    img = row.find("td", class_="posterColumn").find("img").get("src")
    #print(rank, title, link, year, rating, sep=";")
    movie = {
        "rank" : rank, "title" : title, "link" : link, "year" : year,
        "rating" : rating, "img" : img
    }
    result.append(movie)
```

```python
df = pd.DataFrame(result)
df
```

```python
df.link = "https://imdb.com" + df.link
df
```
