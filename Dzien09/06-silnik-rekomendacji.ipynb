```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sentences = ["Julia kocha Marka", "Marek kocha Annę", "Anna i Julia kochają Marka", "Wczoraj padał śnieg"]
```

```python
cvec = CountVectorizer()
wm = cvec.fit_transform(sentences)
```

```python
pd.DataFrame( wm.toarray(), columns=cvec.get_feature_names() )
```

```python
pd.DataFrame( cosine_similarity(wm) )
```

```python
df = pd.read_csv("http://51.91.120.89/extras/movie_dataset.csv")
df.head(3)
```

```python
df.columns
```

```python
# wybór cech do silnika rekomendacji
features = ['genres', 'keywords', 'title', 'cast', 'director' ]
```

```python
df.isna().sum()
```

```python
for col in features:
    df[col].fillna("", inplace=True)
```

```python
# tworzenie kolumny dla potrzeb obliczenia odl. cosinusowej
df["recommend"] = df.genres+" "+df.keywords+" "+df.title+" "+df.cast+" "+df.director
```

```python
df["recommend"].iloc[123]
```

```python
cvec = CountVectorizer()
wm = cvec.fit_transform(df.recommend)
wm.shape
```

```python
# obliczanie podob. cosinusowego
similarity = cosine_similarity(wm)
similarity.shape
```

```python
list(similarity[123])
```
