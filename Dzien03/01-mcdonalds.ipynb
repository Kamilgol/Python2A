```python
# '3.9 oz (111 g)'
def to_gram1(s):
    if not '(' in s or s[-2:]!='g)':
        return s
    return s.split('(')[1].replace(')','')

def to_gram2(s):
    if not 'g)' in s:
        return s
    ix1 = s.find("(")
    if ix1<0: return s
    ix2 = s.find(")")
    if ix2<0: return s
    res = s[ix1+1 : ix2]
    return res
```

```python
%timeit -r 10 -n 2000 df["Serving Size"].apply(to_gram1)
```

```python
%timeit -r 10 -n 2000 df["Serving Size"].apply(to_gram2)
```

```python

import re

def to_gram3(s):
    if not 'g)' in s:
        return s
    return re.search("\((.*)\)", s).group(1)
```

```python
%timeit -r 10 -n 2000 df["Serving Size"].apply(to_gram3)
```
