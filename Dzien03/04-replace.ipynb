```python
# 10 packages
def only_digits(s):
    res = ""
    for ch in s:
        if ch>="0" and ch<="9":
            res += ch
    return res
```

```python
only_digits("7pkgs")
```

```python
%timeit -r 10 -n 1000 df.No_of_pkg.apply(only_digits)
```

```python
%timeit -r 10 -n 1000 df.No_of_pkg.replace("[A-Za-z ]" , "", regex=True)
```
