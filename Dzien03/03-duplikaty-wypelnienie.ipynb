```python
# w komlumnie day zamiana NaN na -
df.day.fillna("-", inplace=True)
```

```python
# w kolumnie temp zamiana NaN na wart. średnią
df.temp.fillna( df.temp.mean() , inplace=True)
```

```python
# zamiana tesktu 'Not available' na NaN
# a potem wypelnienie wart. średnią
df['wind-speed'].replace('Not available', pd.np.NaN, inplace=True)
```

```python
df['wind-speed'] = df['wind-speed'].astype('float')
```

```python
df['wind-speed'].fillna( df['wind-speed'].mean() , inplace=True )
```

```python
df.dates = pd.to_datetime(df.dates, format='%m/%d/%Y')
```
