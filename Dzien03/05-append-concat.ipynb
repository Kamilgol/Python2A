### Concat

```python
#
# A0 B0
# A1 B1
# A2 B2
# A3 B3
df1 = pd.DataFrame({
    "A" : ["A"+str(x) for x in range(4)],
    "B" : ["B"+str(x) for x in range(4)],
    "C" : ["C"+str(x) for x in range(4)],
    "D" : ["D"+str(x) for x in range(4)]
})
df1
```

```python
df2 = pd.DataFrame({
    "A" : ["A"+str(x) for x in range(4,8)],
    "B" : ["B"+str(x) for x in range(4,8)],
    "C" : ["C"+str(x) for x in range(4,8)],
    "D" : ["D"+str(x) for x in range(4,8)]
})
df2
```

```python
pd.concat([df1,df2], axis=0) # łaczenie wierszami
```

```python
pd.concat([df1,df2], axis=1) # łaczenie kolumnami
```

```python
df1 = pd.DataFrame({
    "A" : ["A"+str(x) for x in range(4)],
    "B" : ["B"+str(x) for x in range(4)],
    "C" : ["C"+str(x) for x in range(4)],
    "D" : ["D"+str(x) for x in range(4)]
}, index=[1,2,3,4])


df2 = pd.DataFrame({
    "A" : ["A"+str(x) for x in range(4,8)],
    "B" : ["B"+str(x) for x in range(4,8)],
    "C" : ["C"+str(x) for x in range(4,8)],
    "D" : ["D"+str(x) for x in range(4,8)]
}, index=[4,5,6,7])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2], axis=1) # łaczenie kolumnami
```

```python
pd.concat([df1,df2], axis=1, join='inner')
```
