```python
import numpy as np
import pandas as pd
```

### Dyskretyzacja

```python
np.random.seed(0)
age = np.random.randint(0, 106, 40)
age
```

```python
df = pd.DataFrame(age, columns=["Wiek"])
df
```

```python
# >=0 i <18 - dziecko
# >=18 i <65 - dorosly
# >=65 - senior
df["Kategoria"] =pd.cut(df.Wiek, [0, 18, 65, 105], right=False, labels=["dziecko","dorosły","senior"])
df
```

### One-hot encoding

```python
data = np.random.choice(["A","B","C"], 10)
df = pd.DataFrame(data, columns=["Znak"])
df
```

```python
df1 = pd.concat([df, pd.get_dummies(df.Znak)], axis=1)
df1.set_index("Znak", inplace=True)
df1[ sorted(df1.columns, reverse=True) ]
#df1[ sorted(df.columns, reverse=True)]
```
