```python
df = pd.read_csv("../Dzien02/nba1.csv", sep=";")
df
```

```python
dfg = df.groupby("Team")
```

```python
dfg[ ["Salary"] ].max()
```

```python
dfg[["Age"]].agg([min, max])
```

```python
dfg[["Salary"]].agg([pd.np.mean, pd.np.median])
```

```python
from scipy.stats import trim_mean

def tm(x):
    return trim_mean(x.dropna(), .15)
```

```python
df_res = dfg.agg({
    "Age" : [min,max],
    "Salary" : [pd.np.mean, pd.np.median, tm]
})

df_res
```

```python
df_res["mean_vs_tm"] = df_res[('Salary','mean')] - df_res[('Salary', 'tm')]
df_res
```
