```python
# plik CSV bez wiersza nagłówkowego
_names = ['company', 'rank', 'revenues', 'revenue_change', 'profits', 'assets',
       'profit_change', 'ceo', 'industry', 'sector', 'previous_rank',
       'country', 'hq_location', 'website', 'years_on_global_500_list',
       'employees', 'total_stockholder_equity']

df = pd.read_csv("http://51.91.120.89/extras/f500nh.csv", header=None, names=_names, dtype={
    "rank" : "int16", "revenue_change" : "float32"
})
df.dtypes
```
