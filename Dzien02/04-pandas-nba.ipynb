```python
# usuwam wiersze w których znajdują się wszędzie wartości NaN
df.dropna(axis=0, how='all', inplace=True)
df.isna().sum()
```

```python
# zamiana NaN na łańcuch pusty dla kolumny 'College'
df["College"].replace(np.NaN, "", inplace=True)
df.isna().sum()
```

```python
df
```

```python
# konwersja funt -> kg
df.Weight /= 2.2
df
```

```python
def convert_height(s):
    # 6-10
    feet, inches = s.split("-")
    return int(feet)*30.4 + int(inches)*2.54
```

```python
df.Height = df.Height.apply(convert_height)
df
```

```python
df.to_csv("nba1.csv", sep=";")
```
